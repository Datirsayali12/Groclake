#**Cataloglake**

Cataloglake is a comprehensive infrastructure for catalog warehousing. Using Cataloglake, builders can generate, store, cache index and disseminate ecommerce catalog. You can store all your catalog at one place once and then reuse it multiple times anywhere like sharing it with shopify, building search capabilities and displaying products in your webstore. For example, Cataloglake can be used to create recommendation agents, conversational ecommerce search agents, create smart filters, generate catalog using images only, category and product page listing on webstores and many more.

#**ModelLake**
 It is part of the Groclake infrastructure, providing AI-powered model services for tasks such as translation, text to speech generation, response generation. The API endpoints facilitate seamless integration with various commerce automation workflows.

```python
import os
from google.colab import userdata
from groclake.modellake import Modellake

os.environ['GROCLAKE_API_KEY'] = userdata.get('groclake_api_key')
os.environ['GROCLAKE_ACCOUNT_ID'] = userdata.get('groclake_account_id')
```

#ModelLake  Translate
Handles text translation requests.
Translate text between multiple languages.

```python
translate_request={
                  "text": [
                      "red shirt under Rs 500"
                  ],
                  "source_lang_code": "en",
                  "target_lang_code": "ml",
                  "model": "openai"
              }

text = modellake.translate(translate_request)
print("Translated Text:", text)
```

### ModelLake ChatComplete
Generates conversational responses using AI language models.

```python
chat_completion_request = {
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "which model are you using?"}
    ]
}

text = modellake.chat_complete(chat_completion_request)
print("Response :", text)
```
